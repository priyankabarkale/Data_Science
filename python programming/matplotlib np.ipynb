```python
# Import Data Manipulation Libraries
import pandas as pd
import numpy as np

# Import Data Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns

#Import Logging
import logging
logging.basicConfig(level = logging.INFO,
                    filemode = 'w',
                    filename = "logfile.log",
                    format= "%(asctime)s - %(message)s - %(levelname)s",
                    force = True
                    )


# Import Filter Warning Libaries
import warnings
warnings.filterwarnings('ignore')

# import style
plt.style.use("ggplot")
```

```python
logging.info("Matplotlib session started....")
```

```python
from matplotlib.lines import lineStyles
x = np.arange(1,100,10)
y = 3*x**2

plt.plot(x,y,linestyle = '--',marker = 'p',color = 'purple')   # Line Plot
plt.title("LINE PLOT")
plt.legend("Trend Analysis")
plt.grid()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
from matplotlib import legend
x = np.arange(0,500,10)
y1 = x**20
y2 = x**30

plt.figure(figsize = (15,8))
plt.title("TREND ANALYSIS")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y1,linestyle = "--",color = "red",marker = "*",label = "Financial Terms")
plt.plot(x,y2,linestyle = "--",color = "blue",marker = "o",label = "Financial Terms")
plt.grid()
plt.tight_layout()
plt.legend()
plt.show()
```

```python
from matplotlib import legend
x = np.arange(0,100,5)
y1 = np.exp(0.05*x)
y2 = np.exp(0.03*x)

plt.figure(figsize = (15,8))
plt.title("TREND ANALYSIS")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y1,linestyle = "--",color = "red",marker = "*",label = "Financial Terms")
plt.plot(x,y2,linestyle = "--",color = "green",marker = "o",label = "Financial Terms")
plt.grid()
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Bar plot

country = ["USA","UK","INDIA","JAPAN"]
GDP = [2000,5000,8000,4000]
colors = ["red","green","blue","orange"]

plt.bar(country,GDP, color = colors, label = 'GDP')
plt.title("Country showing GDP values")
plt.xlabel("Country")
plt.ylabel("GDP")
plt.legend()
plt.show()
```

```python
# scatter plot

x = np.linspace(0,1000,50)
y = np.linspace(0,1000,50)

plt.scatter(x,y)
plt.plot(x,y,linestyle = "--",color = "black",label = "Data Trends")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.legend()
plt.show()
```

```python
# plt.subplot()

x = np.linspace(0,200,10)

plt.figure(figsize = (8,5))

plt.subplot(2,1,1)
plt.plot(x,x**2,linestyle = "--",marker = "P",c = "k",label = "Finance Term")

plt.subplot(2,1,2)
plt.plot(x,x**3,linestyle = "--",marker = "v",c = "r",label = "Current Finance TErm")

plt.tight_layout()
plt.legend()
plt.show()
```

```python
# plt.subplot()

x = np.linspace(0,200,10)

plt.figure(figsize = (8,5))

plt.subplot(1,2,1)
plt.plot(x,x**2,linestyle = "--",marker = "P",c = "k",label = "Finance Term")

plt.subplot(1,2,2)
plt.plot(x,x**3,linestyle = "--",marker = "v",c = "r",label = "Current Finance TErm")

plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.contour(np.random.uniform(0,1,(5,5)))
plt.contourf(np.random.uniform(0,1,(5,5)))
plt.title("contour plot showing Dense Distribution")
plt.show()
```

```python
# Pie plot

plt.pie(np.random.uniform(0,1,4),
        autopct = "%0.2f",
        explode = [0.02,0.02,0.02,0.02],
        labels = ["USA","UK","INDIA","JAPAN"])
plt.title("Pie chart showing Distribution")
plt.legend()
plt.show()
```

```python
# Hist plot

plt.hist(np.random.uniform(0,1,100),bins=10,edgecolor = "Black",color = "purple")
plt.title("Histogram showing Distribution")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# box plot

plt.boxplot(np.random.uniform(0,1,100),label= "Box Plot")
plt.title("Box Plot showing Distribution")
plt.legend()
plt.show()
```
