```python
# encapsulation method

class BankAccount:
  # constructor
  def __init__(self,owner,balance=0):
    self.owner=owner      # public attribute/features
    self.__balance=balance  # private attribute/features

# getter method (Read only access to balance)
  def get_balance(self):
    return self.__balance

# Deposit method with validation
  def deposit(self,amount):
    if amount > 0:
      self.__balance += amount   # amount + __balance
      print(f"deposited: {amount}")
    else:
      print("Invalid deposit amount!")

#withdraw methid with validation
  def withdraw(self,amount):
      if 0 < amount <= self. __balance:
        self.__balance -= amount      # amount -__balance
        print(f"Withdraw: {amount}")
      else:
        print("Insufficient funds or invalid amount!")

# using the class ---
acc = BankAccount("Priyanka",1000)

acc.deposit(500)
acc.withdraw(300)

print("Current Balance:", acc.get_balance())
```

```python
# Expamle of Encapsulation

class BankAccount:
    def __init__(self,owner,pin,balance=0):
      self.owner = owner
      self.__pin = pin
      self.__balance = balance

 # Method to verify pin (private  helper method)
    def __verify_pin(self,pin):
      return pin == self.__pin     # comparison

 # public method - cheak balance
    def cheak_balance(self,pin):
      if self.__verify_pin(pin):
        return f"Available balance:{self.__balance}"
      else:
        return "Invalid pin"

# public method - Deposit money
    def deposit(self,pin,amount):
      if not self.__verify_pin(pin):
        return "Invalid pin"

      if amount > 0:
        self.__balance += amount
        return f"Deposoted{amount}. New balance: {self.__balance}"
      return "Invalid deposit amount"

# public method - withdraw money
    def withdraw(self,pin,amount):
      if not self.__verify_pin(pin):
        return"Invalid pin"

      if 0 < amount <= self.__balance:
        self.__balance -= amount
        return f"withdraw {amount}. Reamining balance: {self.__balance}"
      return "insufficient funds or invalid amount."


# test Case:
account = BankAccount("priyanka",pin=1234,balance=5000)

print(account.cheak_balance(1234))
print(account.withdraw(1234,1000))
print(account.deposit(1234,2000))
print(account.cheak_balance(1122))
```

```python
# Number Guessing Game : With Latency Calculation
import time
import random

start = time.perf_counter()

class MovieBooking:
    def __init__(self, movie_name, pin, total_seats=100, price=250):
        self.movie_name = movie_name # Public
        self.__pin = pin            # Private
        self.__seats_book = 0       # Private
        self.__price = price        # Private
        self.total_seats = total_seats

    def __verify_pin(self, pin):
        return pin == self.__pin

    def check_availability(self, pin):
        if self.__verify_pin(pin):
            available = self.total_seats - self.__seats_book
            return f"Available seats: {available}"
        return "Invalid pin"

    def book_tickets(self, pin, tickets):
        if not self.__verify_pin(pin):
            return "Invalid pin"
        if tickets > 0:
            if (self.__seats_book + tickets) <= self.total_seats:
                self.__seats_book += tickets
                total = tickets * self.__price
                return f"Booked {tickets} tickets. Total: {total}"
            return "Not enough seats"
        return "Invalid tickets"

    def cancel_tickets(self, pin, tickets):
        if not self.__verify_pin(pin):
            return "Invalid pin"
        if 0 < tickets <= self.__seats_book:
            self.__seats_book -= tickets
            refund = tickets * self.__price
            return f"Cancelled {tickets} tickets. Refund: {refund}"
        return "Invalid cancellation"

# Test Case: (EXACT SAME STRUCTURE as BankAccount)
show = MovieBooking("Jawan", pin=1234, total_seats=20, price=300)

print(show.check_availability(1234))
print(show.book_tickets(1234, 5))
print(show.book_tickets(1234, 3))
print(show.cancel_tickets(1234, 2))
print(show.check_availability(1122))  # Wrong pin

# Your code that you want to time
end = time.perf_counter() - start
print('{:.6f}s for the calculation'.format(end))
```

```python
# Parent Class
class BankAccount:
  def __init__(self,owner,balance=0):
    self.owner = owner
    self.balance = balance

  def deposite(self,amount):
    self.balance += amount
    print(f"Deposited: {amount}")

  def withdraw(self,amount):
      if amount <=self.balance:
        self.balance -= amount
        print(f"withdraw: {amount}")
      else:
        print(f"Insufficient funds!")

  def get_balance(self):
    return self.balance


class SavingAccount(BankAccount):
  def __init__(self,owner,balance=0,intrest_rate=0.05):
    super().__init__(owner,balance)
    self.intrest_rate = intrest_rate

  def add_intrest(self):
    intrest = self.balance * self.intrest_rate
    self.balance += intrest
    print(f"Intrest added: {intrest}")

class CurrentAccount(BankAccount):
  def __init__(self,owner,balance=0,overdraft_limit=1000):
    super().__init__(owner,balance)
    self.overdraft_limit = overdraft_limit

  def withdraw(self,amount):
    if amount <= self.balance + self.overdraft_limit:
      self.balance -= amount
      print(f"withdraw using overdraft: {amount}")
    else:
      print("Exceeded overdraft limit!")

s = SavingAccount("Priyanka",10000)
s.add_intrest()
print("saving balance:", s.get_balance())

c = CurrentAccount("Priyanka", 5000)
c.withdraw(2500)
print("Current Balance:", c.get_balance())
```

```python
import gradio as gr
import gradio as gr

# ------------------ OOP BACKEND ------------------

class UPIApp:
    def __init__(self, name, pin, balance=0):
        self.name = name
        self.__pin = pin
        self.balance = balance
        self.statement = []
        self.authenticated = False

    def verify_pin(self, entered_pin):
        if entered_pin == self.__pin:
            self.authenticated = True
            return " Access Granted! You may now use all UPI features."
        else:
            return " Wrong PIN. Access Denied."

    def check_balance(self):
        if not self.authenticated:
            return " Please enter correct PIN first."
        return f" Current Balance: {self.balance}"

    def credit(self, amount):
        if not self.authenticated:
            return " Please enter correct PIN first."
        if amount <= 0:
            return " Enter valid amount."
        self.balance += amount
        self.statement.append(f"Credited: {amount}")
        return f" {amount} credited successfully."

    def debit(self, amount):
        if not self.authenticated:
            return " Please enter correct PIN first."
        if amount <= 0:
            return " Enter valid amount."
        if amount > self.balance:
            return " Insufficient balance."
        self.balance -= amount
        self.statement.append(f"Debited: {amount}")
        return f" {amount} debited successfully."

    def show_statement(self):
        if not self.authenticated:
            return " Please enter correct PIN first."
        if not self.statement:
            return " No transactions yet."
        return "\n".join(self.statement)

    def apply_loan(self, amount):
        if not self.authenticated:
            return " Please enter correct PIN first."
        if amount <= 0:
            return " Enter valid amount."
        self.balance += amount
        self.statement.append(f"Loan credited: {amount}")
        return f" Loan of {amount} approved & credited."


# Create an object
user = UPIApp("User1", "1234", 5000)

# ------------------ GRADIO FUNCTIONS ------------------

def verify(pin):
    return user.verify_pin(pin)

def balance():
    return user.check_balance()

def credit(amount):
    return user.credit(amount)

def debit(amount):
    return user.debit(amount)

def statement():
    return user.show_statement()

def loan(amount):
    return user.apply_loan(amount)


# ------------------ GRADIO UI ------------------

with gr.Blocks() as app:
    gr.Markdown("##  UPI Application (OOP + Gradio)")

    with gr.Tab(" Verify PIN"):
        pin = gr.Textbox(label="Enter 4-digit UPI PIN", type="password")
        pin_btn = gr.Button("Verify PIN")
        pin_output = gr.Textbox(label="Result")
        pin_btn.click(verify, inputs=pin, outputs=pin_output)

    with gr.Tab(" Check Balance"):
        bal_btn = gr.Button("Check Balance")
        bal_output = gr.Textbox(label="Balance")
        bal_btn.click(balance, outputs=bal_output)

    with gr.Tab(" Credit"):
        credit_amt = gr.Number(label="Enter amount to credit")
        credit_btn = gr.Button("Credit")
        credit_output = gr.Textbox(label="Result")
        credit_btn.click(credit, inputs=credit_amt, outputs=credit_output)

    with gr.Tab(" Debit"):
        debit_amt = gr.Number(label="Enter amount to debit")
        debit_btn = gr.Button("Debit")
        debit_output = gr.Textbox(label="Result")
        debit_btn.click(debit, inputs=debit_amt, outputs=debit_output)

    with gr.Tab(" Statement"):
        stmt_btn = gr.Button("Show Statement")
        stmt_output = gr.Textbox(label="Mini Statement", lines=8)
        stmt_btn.click(statement, outputs=stmt_output)

    with gr.Tab(" Loan"):
        loan_amt = gr.Number(label="Enter loan amount")
        loan_btn = gr.Button("Apply for Loan")
        loan_output = gr.Textbox(label="Result")
        loan_btn.click(loan, inputs=loan_amt, outputs=loan_output)


app.launch()
```
